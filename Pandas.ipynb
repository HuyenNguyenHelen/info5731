```python
#how do i filter rows of pandas dataframe by column value
```

```python
movies = pd.read_csv("http://bit.ly/imdbratings")
```

```python
boolens=[]
for length in movies.duration:
    if length >= 200:
        boolens.append(True)
    else:
        boolens.append(False)
```

```python
len(boolens)
```

```python
is_long = pd.Series(boolens)
```

```python
is_long.head()
```

```python
is_long = movies.duration >= 200
is_long.head()
```

```python
movies[is_long]["genre"] # true
```

```python
movies.loc[movies.duration >= 200,"genre"]
```
